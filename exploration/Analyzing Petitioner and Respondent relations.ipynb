```python
cases['dead_petitioner'] = cases.Petitioner.apply(lambda x: 'dead by' in x.lower())
cases['bank_petitioner'] = cases.Petitioner.apply(lambda x: 'bank' in x.lower())
cases['police_petitioner'] = cases.Petitioner.apply(lambda x: 'police' in x.lower())
```

```python
cases['petitioner_category'] = cases[[ 'dead_petitioner', 'bank_petitioner', 'police_petitioner']].apply(lambda x: pd.np.argmax(x) if x.sum() > 0 else 'general', axis=1)
```

```python
cases.groupby(['petitioner_category', 'Year']).Id.count().plot('barh', figsize=(15, 15))
```

```python
hearings = hearings[hearings.columns[1:]]
```

```python
hearings
```

```python
hearings_aggregate = pd.DataFrame()

for _, group in hearings.groupby('CaseInformationId'):
    last_date = pd.to_datetime(group.HearingDate).max()
    purpose = group.PurposeOfHearing.value_counts()
    temp = pd.Series()
    temp['last_date'] = last_date
    for index in purpose.index:
        temp[index] = purpose[index]
    hearings_aggregate = hearings_aggregate.append(temp, ignore_index=True)
```
